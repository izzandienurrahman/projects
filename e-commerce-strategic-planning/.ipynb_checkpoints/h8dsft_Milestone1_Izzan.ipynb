## Data Loading
source: https://www.kaggle.com/datasets/lissetteg/ecommerce-dataset

```python
import pandas as pd # import library needed
import missingno as msno # missing data visualization
```

```python
pd.options.display.float_format = '{:,.2f}'.format #add thousand separator for display
```

```python
# add custom missing values
missing_values = ["n/a", "na", "--"]
```

```python
df = pd.read_csv("ecommerce_kaggle.csv",na_values=missing_values)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Sales']=(df['Quantity']*df['UnitPrice']).abs() # absolute value for sales because there are negative value in quantity data
```

```python
df[df['Sales']==df['Sales'].max()]
```

## Data Cleaning

```python
msno.matrix(df) #missing data visualization
```

```python
df.isna().sum() # show how many na data we have
```

```python
df['Description'].unique()
```

There are several data which has valid input but with confusing name and or description, to make exploration easier we just omit these rows.

```python
# drop unintelligible product
df.drop(index=df[df['Description']=='Manual'].index,inplace=True)
df.drop(index=df[df['Description']=='SAMPLES'].index,inplace=True)
df.drop(index=df[df['StockCode']=='B'].index,inplace=True)
df.drop(index=df[df['StockCode']=='DOT'].index,inplace=True)

# drop negative quantity product
df.drop(index=df[df['Quantity']<0].index,inplace=True)

# since description and customer_id data is vital, we drop rows where both of their values are NaN
df2 = df.drop(index=df[df['Description'].isna() & df['CustomerID'].isna().any()].index) 
```

```python
df2.isnull().sum()
```

Out of 541,909 rows, we still have 133,626 empty customer id, so it's around 24% of total data. It's a lot of data and still contains useful information, considering the possibility of checking out as a guest user without signing up I guess it is still a valid data so I will just fill NA values with a placeholder CustomerID that is distinct from already existing ones.

```python
df2[df2['CustomerID']==777] # test unique number
```

Seems 777 can be used as a placeholder for CustomerID

```python
df2.fillna(value=777,inplace=True)
```

```python
# remove leading string description
df2['Description'] = df2['Description'].str.strip('Dotcomgiftshop ')
```

```python
before_remove_duplicates = len(df2)
```

```python
#remove duplicates based on all columns
df_clean = df2.drop_duplicates()
after_remove_duplicates = len(df_clean)
duplicate_rows_removed = before_remove_duplicates - after_remove_duplicates
print(f'{duplicate_rows_removed} duplicate rows have been removed! \nThe Dataset now has {after_remove_duplicates} rows.')
```

```python
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'],format="%m/%d/%Y %H:%M")
```

```python
df_clean.reset_index(drop=True,inplace=True) #reset index
```

```python
# drop small last year's data
df_clean.drop(index=df_clean[df_clean['InvoiceDate']<'2011-01-01'].index,inplace=True)
```

```python
df_clean.isna().sum()
```

At first glance it seems now the data is good but wait...

```python
df_clean[df_clean.Sales==0]
```

There are 1000+ data with zero unit price that we also need to get rid of.

```python
df_clean.drop(index=df_clean[df_clean.Sales==0].index,inplace=True)
```

```python
msno.matrix(df_clean)
```

Straight line on the end of chart meaning there are no more missing value.

```python
df_num_rows = len(df.index)
df_clean_num_rows = len(df_clean.index)
print("Original dataset: {} rows \nCleaned dataset {} rows".format(df_num_rows,df_clean_num_rows))
print("Dataset preservation: {:.2f}%".format(df_clean_num_rows/df_num_rows*100))
```

```python
df_clean.reset_index(drop=True,inplace=True)
```

```python
df_clean.to_csv('clean_ecommerce_m1.csv') #save as csv
```

## Analysis & Visualization

```python
# import library needed
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
from scipy import stats
```

1. Sales over time

```python
# quarterly sales over time
quarterly_sales_ot = df_clean[['Sales','InvoiceDate']].groupby(pd.Grouper(key='InvoiceDate',freq='1Q')).sum().reset_index()
quarterly_sales_ot['percent_diff'] = quarterly_sales_ot.Sales.pct_change()
quarterly_sales_ot
```

```python
xticklabels = ['Q1','Q2','Q3','Q4']
ax = sns.barplot(x=quarterly_sales_ot['InvoiceDate'],y=quarterly_sales_ot['Sales'])
ax.set_xlabel('Quarter')
ax.set_xticklabels(labels=xticklabels)
ax.set_title('Quarterly Sales')
for e in ax.containers:
    ax.bar_label(e, labels=[f'£{x:,.0f}' for x in e.datavalues])
```

In general the trend of sales is always increasing for every quarter.

```python
xticklabels = ['Q1','Q2','Q3','Q4']
ax = sns.barplot(x=quarterly_sales_ot['InvoiceDate'],y=quarterly_sales_ot['percent_diff'])
ax.set_xlabel('Quarter')
ax.set_ylabel('Changes (%)')
ax.set_xticklabels(labels=xticklabels)
ax.set_title('Quarterly Sales Percentage Increases (Q1 as baseline)')
for e in ax.containers:
    ax.bar_label(e, labels=[f'{x*100:,.2f}%' for x in e.datavalues])
```

On average we can expect around 7.46% increases for every quarter.

2. Top 5 profitable product over time

```python

df_top5_profitable_product =df_clean.groupby('StockCode').sum()[['Quantity','Sales']].sort_values('Sales',ascending=False).head(5)\
    .merge(right=df_clean[['StockCode','Description']],how='inner',on='StockCode').drop_duplicates()
df_top5_profitable_product
```

```python
fig, ax = plt.subplots()

sns.barplot(x=df_top5_profitable_product.Description, y=df_top5_profitable_product.Sales, ax=ax)
ax.set_title("Top 5 Profitable Products Over Time")

for i,e in enumerate(ax.containers):
    ax.bar_label(e, labels=[f'£{x:,.0f}' for x in e.datavalues])
ax.set_xlabel("Product Name")  # Naming the x-axis
plt.xticks(rotation=45)
sns.set()
```

The top 5 products that yields the most profits are: paper craft, cakestand 3 tier, party bunting, white t-light holder, cream t-light holder and jumbo bag red retrospot.

3. Most profitable product each month

```python
monthly_sum = df_clean.groupby(by=[pd.Grouper(key='InvoiceDate',freq='1M'),'StockCode']).sum()
```

```python
monthly_sum_sorted_by_sales = monthly_sum.sort_values(by=['InvoiceDate','Sales'],ascending=[True,False])
```

```python
top_sales_monthly_sum = monthly_sum_sorted_by_sales.reset_index().drop_duplicates('InvoiceDate',keep='first')[['InvoiceDate','StockCode','Quantity','Sales']]
top_sales_monthly_sum_detailed = top_sales_monthly_sum.merge(right=df_clean[['StockCode','Description']],how='inner',on='StockCode').drop_duplicates()\
    .sort_values(by=['InvoiceDate','Sales'],ascending=[True,False])
top_sales_monthly_sum_detailed['month'] = top_sales_monthly_sum_detailed['InvoiceDate'].dt.month
top_sales_monthly_sum_detailed['year'] = top_sales_monthly_sum_detailed['InvoiceDate'].dt.year
top_sales_monthly_sum_detailed.month = top_sales_monthly_sum_detailed.month.apply(lambda x: calendar.month_abbr[x])
top_sales_monthly_sum_detailed
```

```python
xtlabels = (top_sales_monthly_sum_detailed.month + " " + top_sales_monthly_sum_detailed.year.astype(str)).to_numpy()
```

```python
fig, ax = plt.subplots()
sns.barplot(x=top_sales_monthly_sum_detailed.InvoiceDate, y=top_sales_monthly_sum_detailed.Sales, ax=ax)
ax.set_title("Highest Product Profitability Per Month")

for i,e in enumerate(ax.containers):
    ax.bar_label(e, labels=[f'{x:,.0f}' for x in e.datavalues])
ax.set_xlabel("Date")  # Naming the x-axis
plt.xticks(rotation=45)
ax.set_ylabel('Total Sales (in Pounds)')
ax.set_xticklabels(xtlabels[1:])
sns.set()
```

Chart above shows the profitability of each month's most salable product. paper craft yields the most profit amongst other popular product.

3. Top sales global

```python
df_top_sales_country = df_clean.groupby(by=[pd.Grouper(key='InvoiceDate',freq='1M'),'Country']).agg({'Quantity':'sum','Sales':'sum'}).reset_index()
df_top_sales_country['month'] = df_top_sales_country['InvoiceDate'].dt.month
df_top_sales_country = df_top_sales_country.groupby('Country').sum().sort_values('Sales',ascending=False).reset_index()
df_top5_sales_country = df_top_sales_country.head()
other_sales = df_top_sales_country['Sales'][6:].sum()
df_top5_plus = df_top5_sales_country.append({'Country':'Others','Sales':other_sales},ignore_index=True)
```

```python
# define Seaborn color palette to use
palette_color = sns.color_palette('bright')
  
# plotting data on chart
plt.pie(df_top5_plus.Sales, labels= df_top5_plus.Country, autopct='%.0f%%')
  
# set title
plt.title("Top Sales Global")

# displaying chart
plt.show()
```

Top 5 countries which have the most sales are: UK, Netherlands, EIRE/Ireland, Germany, and France. UK has 85% of total sales contribution in the last year while the other 4 share nearly the same number around 2-3% and other countries combined contributes to 5% of the total sales.

4. Total transactions for each country

```python
df_transaction_by_country = df_clean.groupby(['Country']).count()['InvoiceNo'].sort_values(ascending=False)
df_transaction_by_country = df_transaction_by_country.to_frame().rename(columns={'InvoiceNo':'num_transactions'}).reset_index()
df_transaction_by_country
```

```python
ax = sns.barplot(x=df_transaction_by_country.head(10).Country,y=df_transaction_by_country.head(10).num_transactions)
plt.xticks(rotation=45)
for i,e in enumerate(ax.containers):
    ax.bar_label(e, labels=[f'{x:,.0f}' for x in e.datavalues])
sns.set()
```

Looks like the vast majority of customers are from United Kingdom.

6. Significance of sales in the last month compared to previous months (confidence level 95%, p-tresh: 0.05)

```python
monthly_sales = df_clean.groupby(by=[pd.Grouper(key='InvoiceDate',freq='1M')]).sum()
avg_first_half_year = monthly_sales.Sales[:6].mean()
avg_first_half_year
```

Average first half year 2011 of is 650,016.79 poundsterling

```python
avg_last_half_year = monthly_sales.Sales[6:].mean()
avg_last_half_year
```

Average last half year 2011 of is 942,097.01 poundsterling

```python
# last month sales significance with respect to last 11 months
t_stat,p_val = stats.ttest_1samp(monthly_sales.Sales[6:], avg_first_half_year)
print('P-value:',p_val/2) #The p-value divided by 2 since the output is two-sided p-value
```

Considering p-tresh 0.05, looks like monthly average sales in the last half year has increased significantly from the first half year.
