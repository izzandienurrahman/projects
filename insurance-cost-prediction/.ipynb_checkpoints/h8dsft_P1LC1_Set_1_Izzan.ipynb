## Outlier Handling

```python
# fungsi untuk membuat histogram dan boxplot dari sebuah fitur
def diagnostic_plots(df, variable):
    # Define figure size
    plt.figure(figsize=(16, 4))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[variable], bins=10)
    plt.title('Histogram')

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(y=df[variable])
    plt.title('Boxplot')

    plt.show()
```

```python
# fungsi untuk mendapatkan upper dan lower boundary dari suatu distribusi
def find_skewed_boundaries(df, variable, distance):
    IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)

    lower_boundary = df[variable].quantile(0.25) - (IQR * distance)
    upper_boundary = df[variable].quantile(0.75) + (IQR * distance)

    return upper_boundary, lower_boundary
```

```python
# plot distribusi data age dan juga boxplotnya untuk melihat outlier
diagnostic_plots(X_train, 'age')
print('\nSkewness Value : ', X_train['age'].skew())
```

Tidak ada outlier pada fitur age.

```python
# plot distribusi data bmi dan juga boxplotnya untuk melihat outlier
diagnostic_plots(X_train, 'bmi')
print('\nSkewness Value : ', X_train['bmi'].skew())
```

Terdapat outlier pada upper limit bmi.

```python
upper_bmi, lower_bmi = find_skewed_boundaries(df=X_train, variable='bmi', distance=1.5)
```

```python
upper_bmi
```

```python
num_bmi_outlier = X_train.bmi[(X_train.bmi > upper_bmi) | (X_train.bmi < lower_bmi)].shape[0]
print(f'Percentage of outlier in bmi is {num_bmi_outlier/X_train.shape[0]*100:.2f}%')
```

Outlier bmi relatif sangat kecil maka data train dapat dilakukan proses trimming, namun karena data sudah terlanjur di split, penulis akan menggunakan teknik capping dengan mereplace value outlier dengan upper limitnya.

```python
X_train.bmi[X_train.bmi > upper_bmi] = upper_bmi # replace outlier di sisi upper dengan upper limitnya
```

```python
# fitur bmi setelah dilakukan capping
diagnostic_plots(X_train, 'bmi')
print('\nSkewness Value : ', X_train['bmi'].skew())
```

Sudah tidak ada outlier pada fitur bmi.

```python
# plot distribusi data age dan juga boxplotnya untuk melihat outlier
diagnostic_plots(X_train, 'children')
print('\nSkewness Value : ', X_train['children'].skew())
```

Tidak ada outlier pada fitur children.

## Scaling

```python
X_train[num_col]
```

Untuk fitur numerikal, penulis akan menggunakan scaling dengan standard scaler sebagai tekniknya, dengan pertimbangan 2 dari 3 fitur terdistribusi secara normal.

```python
# instansiasi scaler
std_scaler= StandardScaler()

# melakukan stardardisasi data train
X_train[num_col]= std_scaler.fit_transform(X_train[num_col])
X_train[num_col].head(5) #menampilkan fitur setelah scaling
```

```python
# melakukan stardardisasi data test
X_test[num_col]= std_scaler.transform(X_test[num_col])
```

## Encoding

```python
X_train[cat_col]
```

Untuk fitur kategorical, penulis akan menggunakan teknik one-hot encoding dengan pertimbangan tidak adanya ordo atau tingkatan dalam fitur kategorikalnya.

```python
enc = OneHotEncoder(sparse=False, handle_unknown='ignore')
X_train_ohe = enc.fit_transform(X_train[cat_col])
X_train_ohe
```

```python
X_train[cat_col].head(3) # crosscheck data train sebelum di encode
```

```python
# X_train setelah di encode
feature_names = enc.get_feature_names_out(input_features=cat_col)
X_train_ohe = pd.DataFrame(X_train_ohe, columns=feature_names)
X_train_ohe.head(3)
```

```python
X_test[cat_col].head(3) # crosscheck data train sebelum di encode
```

```python
# X_test
X_test_ohe = enc.transform(X_test[cat_col]) #transform only
X_test_ohe = pd.DataFrame(X_test_ohe, columns=feature_names)
X_test_ohe.head(3)
```

## Concatenate Numerical and Categorical Features

```python
#reset index data train & test sebelum di concat
X_train.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True) 
```

```python
# X_train numeric only
X_train_num = X_train.drop(columns=cat_col) #drop kolom categorical yang belum di encode

# X_test numeric only
X_test_num = X_test.drop(columns=cat_col) #drop kolom categorical yang belum di encode
```

```python
# X_train numeric + encoded categorical
X_train_final = pd.concat([X_train_num,X_train_ohe],axis=1)

# X_test numeric + encoded categorical
X_test_final = pd.concat([X_test_num,X_test_ohe],axis=1) 
```

```python
X_train_final.head(3) #data train final dengan outlier handling + fit_transform()
```

```python
X_test_final.head(3) #data test final tanpa outlier handling & transform only
```

# Model Definition
Model yang penulis gunakan adalah linear regression untuk memprediksi harga biaya asuransi.<br> 
Kelebihan dari model ini adalah Mudah diimplementasi dan efisien dalam proses training namun kekurangannya adalah sensitif terhadap outlier dan adanya asumsi hubungan yang linier antara variabel dependent dan independent 

```python
# instansiasi model regresi linier
model_lin_reg= LinearRegression() 
```

```python
# konversi ke polynomial regression
from sklearn.preprocessing import PolynomialFeatures
poly_converter = PolynomialFeatures(degree=3, interaction_only=True,include_bias=False)
poly_features = poly_converter.fit_transform(X_train_final)
poly_feature_test = poly_converter.transform(X_test_final)
```

# Model Training

```python
model_lin_reg.fit(poly_features, y_train)
```

# Model Evaluation

Penulis memilih metrik MAE sebagai acuan nilai errornya dan R2 score sebagai representasi dari fit nya sebuah model dengan alasan mudah dibaca dan dipahami.

```python
print('------Train------')
y_train_pred = model_lin_reg.predict(poly_features)
print('MAE:',mean_absolute_error(y_train,y_train_pred))
print("R2 Score:",r2_score(y_train,y_train_pred))
```

```python
print('------Test------')
y_test_pred = model_lin_reg.predict(poly_feature_test)
print('MAE:',mean_absolute_error(y_test,y_test_pred))
print("R2 Score:",r2_score(y_test,y_test_pred))
```

Model mendapatkan error MAE sebesar 4000an untuk kedua train dan test set yang artinya terdapat margin of error sebesar 4000an untuk angka prediksi biaya asuransinya. R2 score yang didapatkan sebesar 0.78 yang artinya model ini merepresentasikan prediksi biaya asuransi sebesar 78%. Penulis merasa model ini masih underfit.
