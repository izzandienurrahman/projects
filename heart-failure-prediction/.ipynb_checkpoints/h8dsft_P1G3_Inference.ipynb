# Import Library

```python
# import library yang dibutuhkan
import pandas as pd
import numpy  as np
import joblib
```

# Load Model

```python
# load model, encoder, scaler, dan selected features
with open('rf_randcv.pkl','rb') as file_1:
    rf_randcv = joblib.load(file_1)

with open('ada_randcv.pkl','rb') as file_2:
    ada_randcv = joblib.load(file_2)

with open('gbc_randcv.pkl','rb') as file_3:
    gbc_randcv = joblib.load(file_3)
```

# Data Loading
## Construct Data Infer

```python
# semua fitur/kolom
features = ['age','anaemia','creatinine_phosphokinase','diabetes','ejection_fraction',\
    'high_blood_pressure','platelets','serum_creatinine','serum_sodium','sex','smoking','time']

np.random.randint(0,9,size=10)
np.random.choice(np.arange(10000,1000000,10000),size=10)

# artificial data infer
age                         = np.random.randint(0,100,size=10)                 
anaemia                     = np.random.randint(0,2,size=10)
creatinine_phosphokinase    = np.random.randint(0,10000,size=10)
diabetes                    = np.random.randint(0,2,size=10)
ejection_fraction           = np.random.randint(0,100,size=10)
high_blood_pressure         = np.random.randint(0,2,size=10)
platelets                   = np.random.randint(2e4,1e6,size=10)
serum_creatinine            = np.random.randint(0.0,1.0,size=10)
serum_sodium                = np.random.randint(100,150,size=10)
sex                         = np.random.randint(0,2,size=10)
smoking                     = np.random.randint(0,2,size=10)
time                        = np.random.randint(1,300,size=10)
   

D = {
    'age':age,
    'anaemia':anaemia,                    
    'creatinine_phosphokinase':creatinine_phosphokinase,
    'diabetes':diabetes,                    
    'ejection_fraction':ejection_fraction,           
    'high_blood_pressure':high_blood_pressure,         
    'platelets':platelets,                   
    'serum_creatinine':serum_creatinine,            
    'serum_sodium':serum_sodium,                
    'sex':sex,                         
    'smoking':smoking,                     
    'time':time,  
    }
data_infer = pd.DataFrame(data=D,columns=features)
```

```python
data_infer.head()
```

# Model Inference

```python
# predict result random forest model
pred_rf = rf_randcv.predict(data_infer)

# predict result adaboost model
pred_ada = ada_randcv.predict(data_infer)

# predict result gradientboost model
pred_gbc = gbc_randcv.predict(data_infer)
```

```python
# show data infer + prediction
data_infer['pred_target_rf'] = pred_rf
data_infer['pred_target_ada'] = pred_ada
data_infer['pred_target_gbc'] = pred_gbc
data_infer
```
