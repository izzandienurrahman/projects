# Import Library

```python
# import library yang dibutuhkan
import pandas as pd
import numpy  as np
import joblib
```

## Load Model

```python
# load model, encoder, scaler, dan selected features
with open('model_log.pkl','rb') as file_1:
    model_log = joblib.load(file_1)

with open('model_svc.pkl','rb') as file_2:
    model_svc = joblib.load(file_2)

with open('model_scaler.pkl','rb') as file_3:
    model_scaler = joblib.load(file_3)

with open('model_encoder.pkl','rb') as file_4:
    model_encoder = joblib.load(file_4)

with open('list_num_cols.txt','rb') as file_5:
    num_col = joblib.load(file_5)

with open('list_scaled_num_cols.txt','rb') as file_6:
    scaled_num_col = joblib.load(file_6)

with open('list_cat_cols.txt','rb') as file_7:
    cat_col = joblib.load(file_7)
```

# Data Loading

## Construct Data Infer

```python
# semua fitur/kolom
features = ['age', 'workclass', 'education_num', 'marital_status', 'occupation',
 'relationship','sex','capital_gain','capital_loss','hours_per_week']

# artificial data infer
age = [39, 40, 28, 50, 40]
workclass = [' Private', ' Self-emp-not-inc', ' State-gov', ' Federal-gov',' Local-gov']
education_num = [13, 12, 10,  7,  8]
marital_status = [' Married-AF-spouse', ' Widowed', ' Divorced', ' Separated',' Widowed']
occupation = [' Craft-repair', ' Transport-moving', ' Sales', ' Adm-clerical', ' Farming-fishing']
relationship = ['southwest', 'southeast', 'northwest', 'northeast', 'southeast']
sex = [' Male', ' Female', ' Male', ' Female', ' Female']
capital_gain = [1471,  2202,  6360,  2174,  4934]
capital_loss = [4356, 1719, 2206,  810, 1762]
hours_per_week = [9, 41, 23, 17, 72]

D = {
    'age':age,
    'workclass':workclass,
    'education_num':education_num, 
    'marital_status':marital_status, 
    'occupation':occupation,
    'relationship':relationship, 
    'sex':sex, 
    'capital_gain':capital_loss, 
    'capital_loss':capital_loss, 
    'hours_per_week':hours_per_week
    }
data_infer = pd.DataFrame(data=D,columns=features)
```

```python
data_infer.head()
```

# Data Inference Preprocessing

## Scaling

```python
#Scaling data inference
df_infer_num= data_infer[num_col].copy()
df_infer_num
```

```python
# transform data numerikal
df_infer_num[scaled_num_col]= model_scaler.transform(df_infer_num[scaled_num_col]) 
df_infer_num
```

## Encoding

```python
#Encoding data inference
data_infer[cat_col]
```

```python
# transform data kategorical
data_infer_cat_encoded = model_encoder.transform(data_infer[cat_col]) # encoding data kategorikal
```

```python
data_infer_cat_encoded
```

```python
feature_names = model_encoder.get_feature_names_out(input_features=cat_col)
df_infer_encoded = pd.DataFrame(data_infer_cat_encoded, columns=feature_names)
df_infer_encoded.head(3)
```

## Concatenate Inference Features
Menggabungkan fitur numerik dan kategorikal yang telah di encode

```python
# menggabungkan data numerikal dan kategorical yang sudah di encode
df_infer_final = pd.concat([df_infer_num,df_infer_encoded],axis=1)
df_infer_final
```

# Inferencing

```python
# Predict data inference dengan model logistik
y_pred_inf_log = model_log.predict(df_infer_final)
y_pred_inf_log
```

```python
# Predict data inference dengan model svc
y_pred_inf_svc = model_svc.predict(df_infer_final)
y_pred_inf_svc
```

```python
df_pred_logistic = data_infer.copy()
df_pred_svc = data_infer.copy()
```

```python
#logistic model inference result
df_pred_logistic['pred_income_bracket'] = y_pred_inf_log
df_pred_logistic.head()
```

```python
# svr model inference result
df_pred_svc['pred_income_bracket'] = y_pred_inf_svc
df_pred_svc.head()
```
