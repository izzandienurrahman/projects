# Import Library

```python
# import library yang dibutuhkan
from tensorflow import keras
import pandas as pd
import numpy  as np
import joblib
```

# Load Model

```python
# load preprocessor
with open('deployment/models/preprocessor.pkl','rb') as file_1:
    preprocessor = joblib.load(file_1)
```

```python
# load ANN model
model = keras.models.load_model('deployment/models/func_model_tuned.hdf5')
```

# Data Loading

```python
data_infer = pd.read_csv('datasets/infer.csv')
```

```python
data_infer.head()
```

# Model Inference

```python
# preprocess data before feedforwarding into the model
preprocessed_data = preprocessor.transform(data_infer)
```

```python
# predict result
predictions = model.predict(preprocessed_data)
```

```python
# convert prediction probability to each appropriate target class
pred = np.where(predictions>0.5,1,0)
```

```python
# show data infer + prediction
data_infer['churn_risk_score'] = pred
data_infer
```
